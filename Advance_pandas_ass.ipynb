Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
def reindex_df(df):
  new_index = range(1, len(df) * 2, 2)

  new_df = df.copy()
  new_df.index = new_index
  return new_df
```

```python
data = {
    'A': [10, 20, 30, 40, 50],
    'B': [100, 200, 300, 400, 500],
    'C': [1000, 2000, 3000, 4000, 5000]
}

df = pd.DataFrame(data)


df_reindexed = reindex_df(df)

print(df_reindexed)
```

Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.

```python
def sum_of_first3(df):
  sum_of_value=0
  count=0

  if len(df['Values']) > 3:
    for i in df['Values']:
      sum_of_value+=i
      count+=1
      if count == 3:
            break

  return sum_of_value


data = {
    'Values': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

sum_of_first3(df)
```

Q5. How are DataFrame.size() and DataFrame.shape() different?


Ans: DataFrame.size() and DataFrame.shape() are both methods in Pandas to obtain information about the dimensions of a DataFrame, but they serve different purposes:

DataFrame.size(): This method returns the total number of elements in the DataFrame. In other words, it returns the product of the number of rows and the number of columns. It includes all elements, including the NaN or missing values.

```python

data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data)


size = df.size
print(size)
```

DataFrame.shape(): This method returns a tuple representing the dimensions of the DataFrame. The tuple contains two elements: the number of rows and the number of columns, respectively.

```python

data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data)

shape = df.shape
print(shape)
```

In summary, the key difference is that DataFrame.size() returns the total number of elements in the DataFrame, while DataFrame.shape() returns a tuple representing the dimensions (rows and columns) of the DataFrame.

Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.

Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
function should return a new DataFrame that contains only the selected rows.


```python
data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}
df = pd.DataFrame(data)

def filter_dataframe(df):
    filtered_df = df[(df['A'] > 5) & (df['B'] < 10)]
    return filtered_df

filtered_df = filter_dataframe(df)


print(filtered_df)
```

Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
median, and standard deviation of the values in the 'Values' column.

```python

data = {'Values': [3, 8, 6, 2, 9]}
df = pd.DataFrame(data)

def calculate_statistics(df):

    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_value = df['Values'].std()

    return mean_value, median_value, std_value


mean_val, median_val, std_val = calculate_statistics(df)


print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_val)
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python

data = {'Sales': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55],
        'Date': pd.date_range(start='2023-08-01', periods=10)}
df = pd.DataFrame(data)

def calculate_moving_average(df, window_size=7):

    df_sorted = df.sort_values(by='Date')

    df_sorted['MovingAverage'] = df_sorted['Sales'].rolling(window=window_size, min_periods=1).mean()

    return df_sorted


df_with_moving_average = calculate_moving_average(df)


print(df_with_moving_average)
```

Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
For example, if df contains the following values:
Date
0 2023-01-01
1 2023-01-02
2 2023-01-03
3 2023-01-04
4 2023-01-05
Your function should create the following DataFrame:


Date Weekday
0 2023-01-01 Sunday
1 2023-01-02 Monday
2 2023-01-03 Tuesday
3 2023-01-04 Wednesday
4 2023-01-05 Thursday
The function should return the modified DataFrame.

```python
import pandas as pd

# Assuming you have the DataFrame 'df' with the 'Date' column
# Example DataFrame
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df = pd.DataFrame(data)

def add_weekday_column(df):
    # Convert the 'Date' column to datetime type
    df['Date'] = pd.to_datetime(df['Date'])

    # Create the 'Weekday' column containing weekday names
    df['Weekday'] = df['Date'].dt.strftime('%A')

    return df

# Call the function to add the 'Weekday' column
df_with_weekday = add_weekday_column(df)

# Display the DataFrame with the new 'Weekday' column
print(df_with_weekday)
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd

# Assuming you have the DataFrame 'df' with the 'Date' column
# Example DataFrame
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-25', '2023-02-05']}
df = pd.DataFrame(data)

def select_rows_between_dates(df, start_date, end_date):
    # Convert the 'Date' column to datetime type
    df['Date'] = pd.to_datetime(df['Date'])

    # Use boolean indexing to select rows within the date range
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

    return selected_rows

# Call the function to select rows within the date range
start_date = '2023-01-01'
end_date = '2023-01-31'
selected_rows_df = select_rows_between_dates(df, start_date, end_date)

# Display the DataFrame with selected rows
print(selected_rows_df)
```
